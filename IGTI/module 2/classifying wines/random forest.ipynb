```python
df = pd.read_csv('winequality-red.csv', sep=';')
```

```python
#in this case, the values of the variable quality_wine were replaced by binaries values, that is, values that 
#quality wine were <= 5, receive value == 0, in other case, receive value == 1. It was made in order to test if
#the classifier would be a better accuracy.

df['quality_bin'] = df['quality']

dic_quality = {6: 1, 7: 1, 8: 1, 5:0 , 4:0, 3:0}


#substituindo os valores na coluna 'MSZoning'
df['quality_bin'].replace(dic_quality,inplace=True)
```

```python
#defining dataset training (dropping the columns quality and quality_bin)
df_train = df.drop(['quality', 'quality_bin'],axis=1)
```

```python
#defining dataset target
target_train = df['quality_bin']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scaled = scaler.fit_transform(df_train)
```

```python
df_train_scaled = pd.DataFrame(df_scaled,columns=df_train.columns)
```

```python
df_train_scaled.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train_scaled, target_train, 
                                                    test_size=0.3, random_state=1)
```

```python
from sklearn import metrics
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_random_forest = RandomForestClassifier(max_depth=10, random_state=1)
```

```python
clf_random_forest.fit(X_train, y_train)
```

```python
y_pred_random_forest = clf_random_forest.predict(X_test)
```

```python
round(metrics.accuracy_score(y_test,y_pred_random_forest),2)
```
